```python
def clean_data(st): #String Cleaning or Whitespace Removal
    st = st.strip()
    return st
```

```python
data['education'] = data['education'].replace(['Graduate', 'Not Graduate'], [1,0]) # value replacement
```

```python
from sklearn.model_selection import train_test_split #importing train_test_split for model evaluation of subset set of data
```

```python
input_data = data.drop(columns=['loan_status'])
output_data = data['loan_status']
```

```python
input_data
```

```python
output_data
```

```python
x_train, x_test, y_train, y_test = train_test_split(input_data, output_data, test_size =0.2) #splitting data for train and test purpose
```

```python
x_train.shape, x_test.shape, y_train.shape, y_test.shape
```

```python
from sklearn.preprocessing import StandardScaler #normalize the features in a dataset
```

```python
scaler = StandardScaler()
```

```python
x_train_scaled = scaler.fit_transform(x_train)
```

```python
x_test_scaled = scaler.transform(x_test)
```

```python
from sklearn.linear_model import LogisticRegression #Importing LogisticRegression for fitting a linear model for prediction because the target variable in this case is categorical (loan approval or rejection)
```

```python
model =  LogisticRegression()
```

```python
model.fit(x_train_scaled, y_train)
```

```python
model.score(x_test_scaled, y_test)
```

```python
# 0.9262295081967213 (or 92.6%) accuracy indicates that your model has good predictive performance on the test set.
# This score suggests that the model is correctly predicting the target (loan approval) with high reliability, assuming the dataset and task are appropriate for this level of accuracy.
```
